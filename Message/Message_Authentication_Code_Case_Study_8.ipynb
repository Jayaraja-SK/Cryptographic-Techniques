- If not, then how about k=64, 96 and 126

### TABLE GENERATION

Each key is paired with a message and each key-message pair is assigned with some tag

```python
import itertools
import random
import math
import prettytable

m_len=int(input("ENTER NO. OF BITS FOR MESSAGE = "))
k_len=int(input("ENTER NO. OF BITS FOR KEY = "))
t_len=int(input("ENTER NO. OF BITS FOR TAG = "))

#ALL POSSIBLE N-BIT NUMBERS ARE GENERATED FOR MESSAGES, KEYS AND TAGS
msg_comb=list(itertools.product([0, 1],repeat=m_len)) 
key_comb=list(itertools.product([0, 1],repeat=k_len))
tag_comb=list(itertools.product([0, 1],repeat=t_len))


d=dict()


for i in msg_comb:
    temp=tag_comb.copy()
    
    for j in key_comb:
        if(len(temp)==0):
            temp=tag_comb.copy()
            
        value=random.choice(temp)
        d[(i,j)]=value
        temp.remove(value)

print()
for i in d:
    print(f'MESSAGE = {i[0]}\t| KEY = {i[1]}\t| TAG = {d[i]}')
```

------------------------------------------------

```python
prob=dict()

for i in key_comb:
    prob[i]=0

for i in d:
    for j in d:
        if(i!=j and i[0]!=j[0] and i[1]==j[1] and d[i]==d[j]):
            prob[i[1]]=prob[i[1]]+1

total=0
            
for i in prob:
    total=total+prob[i]

for i in prob:
    if(total!=0):
        prob[i]=prob[i]/total

for i in prob:
    print(f'KEY = {i}  PROB. = {prob[i]}')
```
